#**1. 인구 피라미드**

#**남자 데이터 정의**

```python
import pandas as pd
#usecols = 가져오는 데이터
df_m = pd.read_excel('[예시]인구피라미드.xlsx', index_col='행정기관', usecols='B,E:Y')

#전국데이터는 .를 없애기
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)
```

#**여자 데이터 정의**

```python
df_w = pd.read_excel('[예시]인구피라미드.xlsx', index_col='행정기관', usecols='B,AB:AV')
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)

#여자columns이 남자와 중복되어 '.1'이 생기므로 통일해줌
df_w.columns = df_m.columns
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='NanumBarunGothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus']=False #깨지는현상
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0]//1000)
plt.barh(df_w.columns, df_w.iloc[0]//1000)
plt.title('2011 대한민국 인구 피라미드')
plt.show()
plt.savefig('2011_인구피라미드.png', dpi=100)
```
